```python
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

# 1. Creation du dataset

```python
# Creation des donnes
true_fun = lambda x: np.cos(1.5 + np.pi * x)

np.random.seed(0)
n_samples = 30

n = [1, 4, 15]

for i in range(len(n)):

    x = np.sort(np.random.rand(n_samples)).reshape(x.shape[0], 1)
    y = true_fun(x) + np.random.randn(n_samples, 1) * 0.1

    poly_features = PolynomialFeatures(degree = n[i], include_bias = False)
    x = poly_features.fit_transform(x)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

    # Le model
    model = LinearRegression()
    model.fit(x_train, y_train)

    # Predictions 
    plt.figure()
    plt.scatter(x[:,0],y)
    plt.plot(x[:,0], true_fun(x[:,0]), label ="Fonction relle")
    plt.plot(x[:,0], model.predict(x), c="red", label=f"Rgression de degres {n[i]}")
    plt.legend()
```
