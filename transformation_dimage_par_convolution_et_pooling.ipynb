# 0. Import des librairies

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import scipy 
import skimage.measure as sm
```

# 1. Import d'une image

```python
img = Image.open("/home/jokos29/Téléchargements/photo_chat.jpg")
imgGray = img.convert("L") # converti TGB -> gray
imgGray = np.array(imgGray)
plt.imshow(imgGray, cmap = "gray")
plt.show()
```

```python
print(f"Dimension de l'image = {imgGray.shape}")
print(imgGray)
```

# 2. Le pooling

```python
pooling_factor = 2
reducted_img = sm.block_reduce(imgGray, (pooling_factor, pooling_factor), np.max)
print(f"Dimension de l'image = {reducted_img.shape}")
plt.imshow(reducted_img, cmap = "gray")
```

# 3. Produits de concolution

```python
from scipy import signal
```

# 3.1 Flou

Chaque pixel est rmeplacer par la moyenne des pixels autour

```python
A = np.copy(imgGray)
M = 1/16*np.ones((4,4))
S = signal.convolve2d(A,M, mode = "same", boundary= "fill")

plt.subplot(1,2,1)
plt.imshow(A, cmap = "gray")

plt.subplot(1,2,2)
plt.imshow(S, cmap = "gray")
```

# 3.2 Detection de contours

```python
A = np.copy(imgGray)
M = 1/16*np.array([[-1,-1,-1],
                 [-1,8,-1],
                 [-1,-1,-1]])

S = signal.convolve2d(A,M, mode = "same", boundary= "fill")

plt.subplot(1,2,1)
plt.imshow(A, cmap = "gray")

plt.subplot(1,2,2)
plt.imshow(S, cmap = "gray")
```

## 3.3Mise en relief de lignes verticales

```python
A = np.copy(imgGray)
M = 1/16*np.array([[-1,2,-1],
                 [-1,2,-1],
                 [-1,2,-1]])

S = signal.convolve2d(A,M, mode = "same", boundary= "fill")

plt.subplot(1,2,1)
plt.imshow(A, cmap = "gray")

plt.subplot(1,2,2)
plt.imshow(S, cmap = "gray")
```

## 3.4 Mise en relief de lignes horizontales

```python
A = np.copy(imgGray)
M = 1/16*np.array([[-1,-1,-1],
                 [2,2,2],
                 [-1,-1,-1]])

S = signal.convolve2d(A,M, mode = "same", boundary= "fill")

plt.subplot(1,2,1)
plt.imshow(A, cmap = "gray")

plt.subplot(1,2,2)
plt.imshow(S, cmap = "gray")
```
